```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

filename1 = 'Spring (Undamped) Raw Data.csv'

with open(filename1) as su:
    data_su = pd.read_csv(su)
    
time_su = data_su['Time (s)']
a_x_su = data_su['Acceleration y (m/s^2)']

plt.plot(time_su, a_x_su)

plt.grid()
plt.title('Acceleration in Undamped Spring Oscillation')
plt.xlabel('Time (s)')
plt.ylabel('Vertical Acceleration ($m/s^2$)')
plt.xlim(152.2,162)
```

```python
filename2 = 'Spring (Damped) Raw Data.csv'

with open(filename2) as sd:
    data_sd = pd.read_csv(sd)
    
time_sd = data_sd['Time (s)']
a_x_sd = data_sd['Acceleration y (m/s^2)']

plt.plot(time_sd, a_x_sd)

plt.grid()
plt.title('Acceleration in Damped Spring Oscillation')
plt.xlabel('Time (s)')
plt.ylabel('Vertical Acceleration ($m/s^2$)')
plt.xlim(125.7,133)
```
